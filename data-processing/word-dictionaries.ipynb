```python
import json
import pandas as pd
from collections import defaultdict
```

```python
df = pd.read_csv("cleaned-dataset.tsv", sep='\t').drop("Unnamed: 0", axis=1)
```

```python
# squad files created in data-processing.ipynb

with open("squads_13_14.json", "r") as f:
  squads_13_14 = json.load(f)

with open("squads_14_15.json", "r") as f:
  squads_14_15 = json.load(f)

with open("squads_15_16.json", "r") as f:
  squads_15_16 = json.load(f)

with open("squads_16_17.json", "r") as f:
  squads_16_17 = json.load(f)

with open("squads_17_18.json", "r") as f:
  squads_17_18 = json.load(f)

season_to_squad_dict = {"13/14":squads_13_14, "14/15":squads_14_15, "15/16":squads_15_16, 
                        "16/17":squads_16_17, "17/18":squads_17_18}
```

```python
season_to_team_wordlist = {}
for season in df["Season"].unique():
  squad_dict = season_to_squad_dict[season]
  season_to_team_wordlist[season] = defaultdict(list)
  wordlist = season_to_team_wordlist[season]
  for team in squad_dict:
    unique_players = list(set(squad_dict[team]))
    for player in unique_players:
      # some player surnames still have multiple words e.g. "Ander Herrera", "De Gea"
      split_name = player.split(' ')
      if len(split_name) > 1:
        for name in split_name:
          wordlist[team].append(name)
      else:
        wordlist[team].append(player)

season_to_team_wordlist
```
